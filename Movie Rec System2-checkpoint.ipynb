```python
## Model Buliding Part
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors=cv.fit_transform(e['tag']).toarray()
```

```python
vectors.shape
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
#skiping NLP
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
e['tag']=e['tag'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
e.iloc[11].title
```

```python
e[e['title']=='Spectre']
```

```python
def recommend(movie):
    index=e[e['title']==movie].index[0]
    distance=similarity[index]
    movie_rec=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movie_rec:
        print(e.iloc[i[0]].title)
```

```python
recommend("Live and Let Die")
```
